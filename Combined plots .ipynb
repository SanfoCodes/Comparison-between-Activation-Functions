```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt

# Input range
x = torch.linspace(-5, 5, 500)

# Activation functions
def relu(x): return F.relu(x)
def sigmoid(x): return torch.sigmoid(x)
def tanh(x): return torch.tanh(x)
def leaky_relu(x): return F.leaky_relu(x, negative_slope=0.01)
def swish(x): return x * torch.sigmoid(x)

# Compute outputs
activations = {
    "ReLU(x)": relu(x),
    "Sigmoid(x)": sigmoid(x),
    "Tanh(x)": tanh(x),
    "LeakyReLU(x)": leaky_relu(x),
    "Swish(x)": swish(x)
}

# Plot
plt.figure(figsize=(10, 6))
for label, y in activations.items():
    plt.plot(x.numpy(), y.numpy(), label=label)

plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.title("Comparison of Activation Functions")
plt.xlabel("Input Value (x)")
plt.ylabel("Activation Output")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt

# Input range
x = torch.linspace(-5, 5, 500, requires_grad=True)

# Activation functions and their derivatives
def relu_derivative(x):
    return (x > 0).float()

def sigmoid_derivative(x):
    sig = torch.sigmoid(x)
    return sig * (1 - sig)

def tanh_derivative(x):
    return 1 - torch.tanh(x) ** 2

def leaky_relu_derivative(x, slope=0.01):
    dx = torch.ones_like(x)
    dx[x < 0] = slope
    return dx

def swish_derivative(x):
    sig = torch.sigmoid(x)
    return sig + x * sig * (1 - sig)

# Compute derivatives
derivatives = {
    "ReLU'(x)": relu_derivative(x),
    "Sigmoid'(x)": sigmoid_derivative(x),
    "Tanh'(x)": tanh_derivative(x),
    "LeakyReLU'(x)": leaky_relu_derivative(x),
    "Swish'(x)": swish_derivative(x)
}

# Plot
plt.figure(figsize=(10, 6))
for label, y in derivatives.items():
    plt.plot(x.detach().numpy(), y.detach().numpy(), label=label)

plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.title("Derivatives of Activation Functions")
plt.xlabel("Input Value (x)")
plt.ylabel("Derivative Value")
plt.grid(True)
plt.legend()
plt.show()
```
